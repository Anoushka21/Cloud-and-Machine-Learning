```python
import matplotlib.pyplot as plt
import numpy as np
```

# V100

```python
x1 = np.linspace(0.01, 17.4, 500)
x2 = np.linspace(17.4, 1000, 500)
y1 = 15.7*1000/17.4 * x1
y2 = np.full_like(x2, 15.7*1000)
plt.figure()
plt.figure(figsize=(8, 6))

plt.plot(x1, y1,color='lightblue')
plt.plot(x2, y2,color='red')
plt.scatter(17.4, 15.7*1000, color='green', zorder=5, label='Transition Point/V100 Limit (17.4, 15700)')
plt.scatter(13.381, 1357.212, color='pink', zorder=5, label='VGG11 - (13.381, 1357.212)')
plt.scatter(15.646, 2141.142, color='orange', zorder=5, label='VGG13 - (15.646, 2141.142)')
plt.scatter(17.641, 2697.852,color='black', zorder=5, label='VGG19 - (17.641, 2697.852)')
intersection_point = 17.4
plt.axvline(x=intersection_point, color='lightblue', linestyle='--',ymax=0.95)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('ARITHMETIC INTENSITY (FLOP/BYTES)')
plt.ylabel('PERFORMANCE (GFLOP/SEC)')
plt.title('Roofline Model for vgg11,vgg13,vgg19 on V100 GPU')
plt.legend(fontsize='small')
plt.savefig('V100.png')
plt.show()
```
